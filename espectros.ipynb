```python
#Representación espectro de una señal

# Genera la señal de prueba y deja el resultado en la variable x
# Parametros
fs  = 1000  # frecuencia de muestreo
f1  = 50    # primera frecuencia
f2  = 120   # segunda frecuencia
dur = 0.4   # Duración 0.4 segundo.

# Vector de tiempos
t = np.arange(0, dur, 1/fs)

# Señal de prueba: x(t) = 0.7·sin(2πf1t) + sin(2πf2t)
x = 0.7 * np.sin(2 * np.pi * f1 * t) + np.sin(2 * np.pi * f2 * t)

#Representación espectro X[k]
X = np.fft.fft(x)     # Guarda el resultado de la FFT aquí
k = np.arange(len(x)) # vector índices

# Quita esto:
# raise NotImplementedError()

plt.plot(k, np.abs(X))
plt.xlabel('k')
plt.ylabel('|X[k]|')



#Representación del espectro en frecuencia normalizada
N = len(x)
fd = k / N   # genera el vector de frecuencia normalizada

# Quita esto:
# raise NotImplementedError()

plt.plot(fd, np.abs(X))
plt.xlabel('$f_d$')
plt.ylabel('|X[k]|')

N = 512
X = np.fft.fft(x, n=N)      # resultado de la FFT con 512 elementos
k = np.arange(N)            # índices 0..N-1
fd = k / N                  # 512 frecuencias discretas normalizadas

# Quita esto:
# raise NotImplementedError()

plt.plot(fd, np.abs(X))
plt.xlabel('$f_d$')
plt.ylabel('|X|')
plt.title('Resultado FFT con $N=512$')

#Representación del espectro centrado en 0
def fftshift(X):
    """Reordena el vector X para centrar el espectro en cero."""
    N = len(X)
    if N % 2 == 0:  # N par
        Xs = np.concatenate((X[N//2:], X[:N//2]))
    else:           # N impar
        Xs = np.concatenate((X[(N-1)//2 + 1:], X[:(N-1)//2 + 1]))
    return Xs

def fftfreq(N):
    '''
    Genera el vector de frecuencias discretas centradas en cero
    para representar el espectro (frecuencia normalizada).
    Hay que distinguir si N es par o impar.
    '''
    if N % 2 == 0:  # N par: [-N/2, ..., N/2-1] / N
        n = np.arange(-N//2, N//2)
    else:           # N impar: [-(N-1)/2, ..., (N-1)/2] / N
        n = np.arange(-(N-1)//2, (N-1)//2 + 1)

    fd = n / N

   
    return fd



N = 1024

X = np.fft.fft(x, n=N)      # espectro con 1024 puntos
Xshift = fftshift(X)        # espectro centrado
fd = fftfreq(N)             # frecuencias centradas en 0


```
