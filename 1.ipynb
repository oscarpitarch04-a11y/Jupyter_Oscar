```python
#Crear un filtro FIR de 5 coeficientes con frecuencia de corte discreta fs = 1 empleando firwin

from scipy.signal import firwin
import numpy as np

my_logger.log("obteniendo coeficientes FIR")

#Obten los coeficientes B con las especificaciones s que se piden:
# YOUR CODE HERE
B = firwin(numtaps=5,cutoff=0.1,fs=1)


print("Los coefficientes obtenidos son:",B)

#Genere una señal x1 que sea un coseno discreto de frecuencia discreta 0.1 y duración 200 muestras , Genere una señal x2 que sea un coseno discreto de frecuencia discreta 0.3 y duración 200 muestras
my_logger.log("generando señales de test")

n = np.arange(200) # vector de tiempo discreto

x1 = np.cos(2*np.pi*0.1*n)
x2 = np.cos(2*np.pi*0.3*n)

#Filtro FIR de un filtro de orden 40 (41 coeficientes) y frecuencia de corte 0.2
from scipy.signal import lfilter 

y1 = lfilter(b=B,a=[1],x=x1)
y2 = lfilter(b=B,a=[1],x=x2)

my_logger.log("programando limit_BW")






from typing import Tuple

def limit_BW(x: np.ndarray, fs: float, band_width: float = 4000.0, num_coef: int = 101) -> Tuple[np.ndarray, np.ndarray, np.ndarray]:
    """
    Aplica un filtro FIR paso bajo a una señal de audio para limitar su ancho de banda.

    Parámetros:
    - x (np.ndarray): Señal de audio de entrada.
    - fs (float): Frecuencia de muestreo de la señal de audio.
    - band_width (float, opcional): Ancho de banda del filtro en Hz. Por defecto es 4000.0 Hz.
    - num_coef (int, opcional): Número de coeficientes del filtro FIR. Por defecto es 101.

    Retorna:
    - np.ndarray: Señal de audio filtrada.
    - np.ndarray: Frecuencias de la respuesta en frecuencia del filtro.
    - np.ndarray: Respuesta en frecuencia del filtro (compleja).
    
    La función realiza las siguientes operaciones:
    1. Calcula los coeficientes de un filtro FIR paso bajo utilizando la función `firwin`.
    2. Aplica el filtro a la señal de audio `x`.
    3. Calcula la respuesta en frecuencia del filtro utilizando la función `freqz`.
    4. Visualiza la respuesta en frecuencia del filtro y el espectro de potencia de la señal filtrada.
    """
    
    # Calcula los coeficientes del filtro FIR paso bajo con la función firwin
    # y aplica el filtro a la señal de audio x
    # Num coef es el número de coeficientes del filtro FIR
    
    B = firwin(numtaps=num_coef,cutoff=band_width,fs=fs)
    y = lfilter(b=B,a=[1],x=x)
    
    # Ahora obtenemos la respuesta en frecuencia del filtro  con la función freqz
    # Pasa el argumento fs=fs para obtener la respuesta en frecuencia en Hz
    
    f,H = freqz(b=B,a=[1],fs=fs)
    

    display(Audio(y, rate=fs))    
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6))

    # Plot the filter frequency response    
    ax1.plot(f, np.abs(H))
    ax1.set_title('Frequency Response of the Filter')
    ax1.set_xlim(0,fs/2)
    ax1.set_xlabel('fd')
    _ = plt.ylabel(r'$|H(e^{j\omega})|$')
    

    # Plot the frequency spectrum
    ax2.psd(y, Fs=fs, NFFT=256)
    ax2.set_xlabel('Frequency (Hz)')
    ax2.set_ylabel('Power Spectral Density (dB/Hz)')
    ax2.set_title('Power Spectrum after filtering')
    ax2.set_xlim(0,fs/2)

    plt.tight_layout()
    plt.show()

    
    return y, f, H
```
