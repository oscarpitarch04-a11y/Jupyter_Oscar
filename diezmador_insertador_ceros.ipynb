```python
#crear insertador de ceros

my_logger.log("programando insertador ceros")

import numpy as np
def insertador_ceros(x: np.ndarray, L: int) -> np.ndarray:
    """Inserta L-1 ceros entre muestras de x (upsampling por factor L)."""
    assert L > 0, 'L debe ser un entero positivo'

    x = np.asarray(x).ravel()   # asegurar 1D
    if L == 1 or x.size == 0:
        return x

    y = np.zeros(L * x.size, dtype=x.dtype)
    y[::L] = x
    return y

x = np.random.randn(7) # vector unidimensional
y = insertador_ceros(x,2)

np.set_printoptions(precision=2)
print(x)
print(y)

#Crear un diezmador

my_logger.log("programando diezmador")

def diezmador(x: np.ndarray, M: int) -> np.ndarray:
    """Reduce la tasa de muestreo de un vector unidimensional (decimaciÃ³n por M)."""
    assert M > 0, 'M debe ser un entero positivo'

    x = np.asarray(x).ravel()  # asegurar que es 1D

    if M == 1 or x.size == 0:
        return x

    y = x[::M]
    return y

x = np.arange(10)
M = 3
print("Entrada ",x)
y = diezmador(x,M)
print("Salida", y, "deberia ser igual a: [0 3 6 9]" )


```
