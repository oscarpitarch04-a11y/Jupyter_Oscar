```python
def nextpow2(N):
    Nout = None  # aqui el resultado

    # YOUR CODE HERE
    Nout = 2 ** int(np.ceil(np.log2(N)))

    # el resultado debe ser entero
    Nout = int(Nout)
    return Nout

def determina_frecuencia_tono(x, fs):
    ''' Devuelve la frecuencia de un tono contaminado por ruido en Hz '''
    # 1. Calcula la longitud de x y el valor de N para la FFT
    L = len(x)
    N = nextpow2(L)

    # 2. Calcula la FFT de x
    X = np.fft.fft(x, N)

    # 3. Busca el índice del máximo del módulo de la FFT en la primera mitad de X
    Xpos = X[:N//2]                 # solo frecuencias positivas
    kmax = np.argmax(np.abs(Xpos))  # índice del pico

    # 4. Calcula la frecuencia discreta del índice kmax
    fd = kmax / N

    # 5. Convierte a Hz empleando fs
    fpico = fs * fd

    return fpico

def determina_frecuencia_rango(x, fmin, fmax, fs):
    ''' Devuelve la frecuencia de un tono en el rango de frecuencias fmin, fmax'''
    
    L = len(x)
    N = nextpow2(L)

    # 2. FFT de x
    X = np.fft.fft(x, N)

    # 3. Frecuencias positivas normalizadas
    fdpos = np.arange(N//2) / N

    # 4. Convertir a Hz
    fpos = fs * fdpos

    # 5. Índices más cercanos a fmin y fmax
    kmin = int(round(fmin / fs * N))
    kmax = int(round(fmax / fs * N))

    # 6. Trozo del espectro en el rango
    Xslice = X[kmin:kmax+1]

    # 7. Índice del máximo dentro del trozo
    k_local = np.argmax(np.abs(Xslice))

    # 8. Índice real dentro de todo X
    k_peak = kmin + k_local

    # 9. Frecuencia estimada
    fpico = (k_peak / N) * fs

    return fpico
```
