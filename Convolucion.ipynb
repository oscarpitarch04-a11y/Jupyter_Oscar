```python
def nextpow2(N):
    '''Devuelve el siguiente valor que es potencia de dos. Por ejemplo si N=57, deberia devolver 64.'''
    
    # Siguiente potencia de 2:
    Nout = 2 ** int(np.ceil(np.log2(N)))

    # El resultado debe ser entero
    Nout = int(Nout)
    return Nout

def convfft(x,h):
    ''' Calcula convolución lineal empleando fft.
        Calcula el valor mínimo de N
        y busca el siguiente valor potencia entera de 2
    '''

    Lx = len(x)
    Lh = len(h)
    L  = Lx + Lh - 1          # longitud de la convolución lineal

    # Calcula el valor mínimo de N y busca el siguiente valor potencia de 2
    Nmin = L
    N = nextpow2(Nmin)

    # Calcula y como fft inversa del producto de ffts de x y h
    X = np.fft.fft(x, N)
    H = np.fft.fft(h, N)
    y = np.fft.ifft(X * H)

    # Quédate con los Lx+Lh-1 primeros elementos de y
    ytrim = y[:L]

    return ytrim

```
