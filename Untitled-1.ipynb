# Diseño de filtros FIR 

En este ejercicio aprenderemos a obtener los coeficientes de filtros FIR empleando la libreria `scipy.signal`

Podemos considerar que los filtros FIR son un caso particular de los filtros IIR en los que no hay parte recurrente. De esta forma la 
salida del filtro se puede escribir como:

$$ y[n] = \sum_{k=0}^K b_k x[n-k] $$

La expresión anterior no es más que la convolución de las señals $x[n]$ y $b[n]$, siendo $b[n] = \sum_{k=0}^K b_k \delta [n-k]$  . Por este motivo, en el caso de los **filtros FIR la respuesta impulsional del filtro $h[n]$ coincide con los coeficientes $b[n]$** y por tanto su duración es finita (de ahi el nombre).

En la siguiente figura podemos ver gráficamente la ecuación anterior y las muestras empleadas para el cálculo de $y[n]$:

![iir_samples](figs/fir_samples.png)


En este caso la función de transferencia del filtro viene caracterizada unicamente por los coeficientes $b_k$ 
y por tanto el proceso de diseño del filtro consistirá en determinar: 
1. Los coeficientes del numerador de $H(z)$:`B = [b0, b1, ... , bK]`

Una de las ventajas principales de los filtros FIR es que nunca se pueden hacer inestables ya que no tienen ningun tipo de realimentación.

Por contra, el número de operaciones requeridas suele ser más elevado en comparación con un filtro IIR equivalente.

### Determinación del orden del filtro

En los ejercicios anteriores el orden del filtro FIR venia como especificación del filtro. En el caso de los filtros FIR el orden del filtro afecta especialmente al ancho de la banda de transición de forma que necesitaremos mayores ordenes para anchos de transición más estrechos (y por tanto más operaciones por muestra).

Como primera aproximación para determinar el orden del filtro haremos lo siguiente. 
1. Normalizaremos el ancho de banda de transición por la frecuencia de muestreo obteniendo $\Delta f = \frac{\mbox{ancho de banda transicion}}{f_s}$
2. Dependiendo del tipo de ventana emplearemos las siguientes reglas para obtener el orden del filtro $N$:
   * Si la ventana es rectangular: $N \ge \frac{2}{\Delta f}$
   * Si la ventana es hamming: $N \ge \frac{3.3}{\Delta f}$
   * Si la ventana es Blackman: $N \ge \frac{5.5}{\Delta f}$

Vemos que la ventana Blackman que era la que tenia una mayor atenuación en la banda rechazada es la que requiere un orden mayor manteniendo el mismo valor de $\Delta f$.

Hay que tener en cuenta que la regla anterior nos debe servir **solo como primera approximación**, siempre debemos representar la respuesta en frecuencia y en caso de que no se ajustara perfectamente subir el orden hasta conseguir cumplir con nuestro objetivo.

Teniendo en cuenta la regla anterior vamos a programar una función que nos devuelva el numero de coeficientes (*taps*) del filtro. **Recuerda que el número de coeficientes es $N+1$**


```python
def firwin_ntaps(bw_transicion: float, window: str = 'hamming', fs: float = 1.0) -> int:
    """ Funcion que calcula el número de coeficicientes para un filtro FIR 
    dependiendo del tipo de ventana.
    NOTA: Redondear al entero mas alto para obtener un valor entero de numero de coeficientes.
    """
    valid_windows = ['boxcar', 'hamming', 'blackman']
    assert window in valid_windows, f"{window} es un tipo de ventana no valido"
    
    # Normalización Δf = BW / fs
    df = bw_transicion / fs

    # Cálculo del orden N según la ventana
    if window == 'boxcar':          # rectangular
        N = 2 / df
    elif window == 'hamming':
        N = 3.3 / df
    elif window == 'blackman':
        N = 5.5 / df

    # Número de coeficientes: ntaps = N + 1
    ntaps = int(np.ceil(N)) + 1

    return ntaps

```

```python
#comprobacion
assert isinstance(firwin_ntaps(100,'hamming', 8000), int), "Recuerda redondear al alza y emplear luego int para convertir a entero"

assert firwin_ntaps(100,'hamming', 8000)==265, "revisa la formula"
assert firwin_ntaps(100,'boxcar', 8000)==161, "revisa la formula"
assert firwin_ntaps(100,'blackman', 8000)==441, "revisa la formula"

my_logger.success("Orden filtro fir metodo ventana", 1.0)
```

```python
from scipy.signal import remez # funcion que usaremos en este apartado

# --- 1. Definición de parámetros ---
fs = 16000.0            # Frecuencia de muestreo
order = 80
numtaps = order + 1     # El número de coeficientes siempre es orden + 1

# Definición de la geometría de las bandas basada en la nota "centrada en fc"
cutoff = 2000.0
trans_width = 150.0

# Calculamos los bordes centrados en 2000 Hz
# Banda Stop termina en: 2000 - 75 = 1925 Hz
# Banda Paso empieza en: 2000 + 75 = 2075 Hz
f_stop_end = cutoff - trans_width / 2
f_pass_start = cutoff + trans_width / 2

bands = [0, f_stop_end, f_pass_start, fs/2] # [0, 1925, 2075, 8000]

# Ganancias deseadas: 0 para Stop, 1 para Paso
desired = [0, 1]

# --- 2. Diseño del filtro con Remez ---
B_highpass = remez(numtaps, bands, desired, fs=fs)

# --- 3. Respuesta en frecuencia ---
# IMPORTANTE: No definir worN o dejarlo por defecto (512) para pasar el assert del test
W_highpass, H_highpass = freqz(B_highpass, 1, fs=fs)

# YOUR CODE HERE
# raise NotImplementedError()

plot_response(W_highpass, H_highpass, f"High-pass Filter (order=80)")
plt.show()
```
