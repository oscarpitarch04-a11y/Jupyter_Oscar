```python
#busqueda de maximo absoluto
t = np.linspace(-5,5,1000)
x = np.sinc(t+2) # sinc adelantada

plt.axes().set_aspect(3)
plt.plot(t,x)
plt.xlabel('tiempo')
plt.ylabel('amplitud')
plt.grid()

xmax = np.max(x)
tmax = t[np.argmax(x)]

Imagina ahora que queremos buscar el máximo de la señal x pero unicamente entre los instantes t = 0, t =2
.

En este caso debemos:

Extraer los fragmentos de x y t que cumplan la condicion (ver cuaderno anterior)
Buscar el máximo en los fragmentos tal y como hemos hecho en el apartado anterior

def find_max_in_range(x: np.ndarray, t: np.ndarray, tmin: float, tmax: float) -> tuple:
    """Encuentra el valor máximo y el instante donde ocurre dentro de un rango especificado."""

    # Extrae los fragmentos correspondientes de t y x que cumplen la condicion
    mask = (t >= tmin) & (t <= tmax)   # Máscara booleana del rango temporal
    x_selected = x[mask]
    t_selected = t[mask]

    # Busca el máximo y su posición en x_selected y t_selected
    x_localmax = np.max(x_selected)
    t_localmax = t_selected[np.argmax(x_selected)]

    return t_localmax, x_localmax

```
