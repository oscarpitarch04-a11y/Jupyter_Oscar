```python
#En este ejercicio vamos a hacer una función para comprobar sin un fichero .wav, cumple con unas ciertas especifiaciones. En concreto pretendemos verificar:

#El número de canales (mono o estereo)
#La amplitud máxima en valor absoluto
#La duración en segundos
#La frecuencia de muestreo
#En el ejercicio emplearemos la función librosa.load para leer los audios. Si consultamos la documentación de la función indica que por defecto el parametro mono=True, esto quiere decir que en caso de que el audio sea estereo realiza la conversión automática a mono. Por tanto para leer el fichero original deberiamos poner: mono=False en la función

#Lea los audios:

#./signals/file_mono.wav
#./signals/file_stereo.wav
import numpy as np
import librosa
from IPython.display import Audio

audio_mono, fs_mono = librosa.load('./signals/file_mono.wav',sr=None, mono=False,) 
audio_stereo, fs_stereo = librosa.load('./signals/file_stereo.wav',sr=None, mono=False)





print(f'audio_mono: {audio_mono.shape}, fs_mono: {fs_mono}')
print(f'audio_stereo: {audio_stereo.shape}, fs_stereo: {fs_stereo}')

#A la vista del resultado anterior escriba una función que devuelva el número de canales:

#1: para mono
#2: para estero de un fichero wav

def num_channels(x: np.array) -> int:
    "Esta función devuelve el número de canales de un audio"
    "x es un array de numpy que representa un audio obtenido tras leer un archivo wav con librosa.load"
    
    # YOUR CODE HERE
    if x.ndim == 1:
        num_channels = 1
    else:
        num_channels = x.shape[0]
    
    return num_channels


#Ahora escriba una función que compruebe lo siguiente:

#Que la frecuencia de muestreo sea sr (parametro de la función)
#Que (`parámetros de la función)
#Que la duración en segundos esté en el rango [dur_min, dur_max], siendo valores parámetros de la función. Puede realizar esta comprobación conociendo el número de muestras de la señal y fs o bien con la función librosa.get_duration
#En caso de que no se cumpla alguna condición anterior o que no pueda leer el audio, la función devolverá False

def check_wav(wav_filename: str, 
        amax: float = 0.9, amin: float = 0.2, 
        dur_min: float = 1.0, dur_max: float = 4.0, 
        sr: int = 22050) -> bool:
    """ Esta función verifica si un archivo wav cumple las siguientes especificaciones:
        - Fichero de audio mono 
        - dur_min: duración mínima permitida
        - dur_max: duración máxima permitida
        - sr: frecuencia de muestreo esperada
        - amax: amplitud máxima permitida
        - amin: amplitud mínima permitida
    """    

    valid = True

    try:
        
        audio, fs = librosa.load(wav_filename, mono=False, sr=None)
    except Exception:
        valid = False
        return valid

    if audio.ndim != 1:
        valid = False

    if fs != sr:
        valid = False

    amp_max = np.max(np.abs(audio))
    if amp_max > amax or amp_max < amin:
        valid = False

    dur = librosa.get_duration(y=audio, sr=fs)
    if dur < dur_min or dur > dur_max:
        valid = False

    return valid

```
